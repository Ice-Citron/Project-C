1. THE BITWISE SANDBOX (Q1 & Q2)
   In C, you cannot index individual bits like an array. To read or write a 
   specific field (like a 4-bit register ID inside a 32-bit instruction), you
   have to mathematically isolate it.
   - EXTRACTING (The Emulator): You move the target bits to the very bottom 
     using a right-shift (`>>`), then you take a "mask" of 1s (like `0b1111`)
     and use bitwise AND (`&`) to chop off everything above it.
   - SETTING (The Assembler): The exact reverse. You clear the target zone in
     the raw instruction using an inverted mask `~` , shift your new data up to 
     the correct position, and drop it in using bitwise OR (`|`).
