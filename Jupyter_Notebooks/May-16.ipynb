CONCEPT: THE C DYNAMIC ARRAY
   In higher-level language, a dynamic array (like Kotlin's `ArrayList`) hides
   its memory management. In C, you must manually coordinate three things: 
   - a pointer to a heap-allocated block of memory, 
   - the number of slots currently containing valid data, 
   - the total number of slots allocated. 
   
   When the array fills up, you must manually ask the OS for a 
   - larger contiguous block, 
   - move the data,
   - update your pointer.

---
1. `size_t`
   This is an unsigned integer type specifically designed for memory and sizes.
   - WHY NOT JUST USE `int`? An `int` can be negative, but an array cannot have
     a negative length. Additionally, on a 64-bit system, an `int` might max out
     at 2 billion, but `size_t` is guaranteed to be large enough to hold the
     maximum possible size of an object or array in your specific system's 
     memory. Always use `size_t` for lengths, capacities, and array indices.

2. `struct`
   This defines a custom block of memory. In our case, an anonymous struct that
   contains a pointer to the heap (`data`), the number of valid items (`len`),
   and the total allocated slots (`cap`).

3. `typedef`
   This tells the C compiler, "Take this anonymous struct we just defined and 
   leet me refer to it using the name `IntVec`."
